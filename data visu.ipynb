```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#count plot
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller type')
plt.ylabel('count')
plt.show()
```

```python
#histogram
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True,color='green')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
#piechart
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars,labels=['First Owner','Second Owner','Third Owner','Fourth & Above Owner','Test Drive Car'],
        autopct='%1.1f%%',colors=['lightcoral','lightgreen','red','blue','pink'],
        startangle=180,explode=(0.05,0.05,0.05,0.05,0.05))
plt.title("Car Details")
plt.show()
```

```python
cars=df['owner'].value_counts()
cars
```

```python

#box plot
plt.figure(figsize=(8,6))
sns.boxplot(x='seller_type',y='selling_price',data=df,palette='viridis')
plt.title('Fare Distribution by passenger class')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(10,8))
sns.scatterplot(x='year',y='km_driven',hue='owner',style= "seller_type",data=df,palette='rainbow')
plt.title('year vs km_driven with owner and seller_type')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='about Car')
plt.show()
```

```python
num=df[['year','selling_price','km_driven']].corr()
num
```

```python
#heatmap
plt.figure(figsize=(10,8))
sns.heatmap(num, annot=True,cmap='rainbow',center=0)
plt.title('correlation heatmap of car dataset')
plt.show()
```

```python
#violinplot
plt.Figure(figsize=(10,8))
sns.violinplot(x='fuel',y='year',hue='seller_type',data=df,
               split=True,palette='Set2')
plt.title('seller_type Distribution by year and fuel')
plt.xlabel('fuel')
plt.ylabel('year')
plt.legend(title='Survived')
plt.show()
```

```python
#kde plot
sns.kdeplot(data=df['year'].dropna(),color='skyblue',shade='')
plt.title('year distribution (kernal Density Estimate)')
plt.xlabel('year')
plt.ylabel('Density')
plt.show()
```

```python

g=sns.FacetGrid(df,col='seller_type',row='owner',margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='teal')
g.fig.suptitle('age Distribution by Class and Survival',y=1.03)
plt.show()
```

```python
#pairplot
sns.pairplot(df[['year','selling_price','km_driven','transmission']],
             hue='transmission',palette='husl',
             plot_kws={'alpha':0.6},diag_kind='kde')
plt.suptitle('Pair plot of titanic variabls',y=1.02)
plt.show()
```

```python
#swarm plot
plt.figure(figsize=(5,5))
sns.swarmplot(x='fuel',y='km_driven',hue='transmission',data=df,palette='Set1',dodge=True)
plt.title("Transmission Distraction by Fuel and km_driven (swarm plot)")
plt.xlabel("fuel")
plt.ylabel("km_driven")
plt.legend(title="transmission")
plt.show()
```
