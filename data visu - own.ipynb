```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='rating',data=df,palette='rainbow')
plt.title('Netflix rating Count')
plt.xlabel('rating')
plt.ylabel('count')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.histplot(df['release_year'].dropna(),bins=30,kde=True,color='green')
plt.title('Hisplot for netflix release year')
plt.xlabel('release_year')
plt.ylabel('count')
plt.show()
```

```python
movies=df['release_year'].value_counts().head(5)
plt.figure(figsize=(8,6))
plt.pie(movies,labels=['2020','2021','2021','2021','2021'],
        autopct='%1.1f%%',colors=['lightcoral','lightgreen','red','blue','pink'],
        startangle=180,explode=(0.05,0.05,0.05,0.05,0.05))
plt.title("Movie release year")
plt.show()
```

```python
movies=df['release_year'].value_counts()
movies
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='rating',y='release_year',data=df,palette='rainbow')
plt.title('Rating and release_year')
plt.xlabel('Rating')
plt.ylabel('release_year')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='release_year',y='rating',hue='rating',style= "release_year",data=df,palette='rainbow')
plt.title('release_year with country')
plt.xlabel('release_year')
plt.ylabel('cast')
plt.legend(title='about movie')
plt.show()
```

```python
de=pd.read_csv(r"D:\DATASETS FOR PROJECT\ML P3 titanic\train.csv")
de.head()
sample_data=de.sample(50)
```

```python
de.info()
de.describe()
de.shape
```

```python
num=de[['PassengerId','Pclass','Age','Fare']].corr()
num
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num,annot=True,cmap='rainbow',center=0)
plt.title('correlation heatmap for passenger')
plt.show()
```

```python

plt.figure(figsize=(10,8))
sns.violinplot(x='Parch',y='Age',hue='Fare',data=sample_data,split=True,palette='Set1')
plt.title('passenger survival distribution')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.legend(title='Survived')
plt.show()
```

```python
sns.kdeplot(data=de['Age'].dropna(),color='green',shade='')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.FacetGrid(sample_data,col='Fare',row='Age',margin_titles=True)
plt.map(sns.histplot,'Parch',bins=20,color='blue')
plt.fig.suptitle('age Distribution',y=1.03)
plt.show()
```
